# Agent Comparison
Load results JSON and display summary table and violin plots.

```python
import json
import glob
import pandas as pd
import plotly.graph_objects as go
from pathlib import Path

# Load latest results
results_files = sorted(glob.glob('../evaluation/results/*.json'))
if not results_files:
    print('No results files found. Run batch_evaluator first.')
else:
    with open(results_files[-1]) as f:
        data = json.load(f)
    summary = data['summary']
    df = pd.DataFrame(summary).T
    print(df[['mean_IS_bps', 'std_IS_bps', 'win_rate_vs_TWAP', 'p_value_vs_TWAP']])
```

```python
# Violin plot of IS distributions
if results_files:
    fig = go.Figure()
    for agent_name, is_vals in data['is_distributions'].items():
        fig.add_trace(go.Violin(y=is_vals, name=agent_name, box_visible=True))
    fig.update_layout(title='IS Distribution by Agent', yaxis_title='IS (bps)')
    fig.show()
```
