# Bandit Learning Curves
Plot IS vs episode number with rolling mean for LinUCB and Thompson.

```python
import json
import glob
import numpy as np
import pandas as pd
import plotly.graph_objects as go

results_files = sorted(glob.glob('../evaluation/results/*.json'))
if not results_files:
    print('No results files found.')
else:
    with open(results_files[-1]) as f:
        data = json.load(f)

    fig = go.Figure()
    for agent_name in ['LinUCB', 'Thompson']:
        if agent_name in data['is_distributions']:
            is_vals = data['is_distributions'][agent_name]
            rolling = pd.Series(is_vals).rolling(window=10).mean()
            fig.add_trace(go.Scatter(
                y=rolling, mode='lines', name=f'{agent_name} (rolling mean 10)'
            ))
    fig.update_layout(
        title='Bandit Learning Curves (Rolling IS)', 
        xaxis_title='Episode', yaxis_title='IS (bps)'
    )
    fig.show()
```
